```
import pandas as pd
all_months1 = pd.read_csv('/content/all_months1.csv')
```

```
all_months1.tail()
```

##CLEANING

```
#drop rows of NaN
Nan_df = all_months1[all_months1.isna().any(axis =1)]
Nan_df.head()
```

```
#this means we have rows with Nans
all_months1 = all_months1.dropna(how= 'all')
#We use all instead of any because any will lose many of our data
all_months1.head()
```

```
#Find 'Or' and delete it
all_months1 = all_months1[all_months1['Order Date'].str[0:2] !='Or']
all_months1.head(1)
```

```
# Change data from string to int
all_months1['Quantity Ordered'] = pd.to_numeric(all_months1['Quantity Ordered'])
all_months1['Price Each'] = pd.to_numeric(all_months1['Price Each'])
```

#Augument data with additional Data

```
#Add month columns
all_months1['month'] =all_months1['Order Date'].str[0:2]
#we need to convert the new created column from string
all_months1['month']= all_months1['month'].astype('int32')
all_months1.head()
```

```
##Add a sales columns

all_months1['sales'] = all_months1['Quantity Ordered'] * all_months1['Price Each']
all_months1.head(5)
```

Task 4: Add a city column

```
#Add a city column
# use the .apply method - it allows us to run any function in our datframe
#we use split

#all_months1['city'] = all_months1['Purchase Address'].apply(lambda x:x.split(',')[1])
#all_months1.head()
#or we can use a function to solve this
def get_city(address):
  return address.split(',')[1]
def get_state(address):
  return address.split(',')[2].split(' ')[1]
  #then in the function, we replace the lambda to the function
all_months1['city'] = all_months1['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")  
all_months1.head()

  
     
```

```
all_months1.drop(columns= 'column', inplace = True)
#all_months1.head()
```

###Question 4. awhat product are most often sold together.


```
#For this we have to check for the Order ID , if there are duplicates then those items are bought o the same order.
df = all_months1[all_months1['Order ID'].duplicated(keep = False)] 
df.head(20)

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df = df[['Order ID','Grouped']].drop_duplicates()
```

```
df.head(100)
```

```
from itertools import combinations
from collections import Counter

count = Counter()
for row in df['Grouped']:
  row_list = row.split(',')
  count.update(Counter(combinations(row_list, 2)))

 # print(count)

  #though this is pretty messy

# to get a more readable list
for key, value in count.most_common(10):
  print(key,value)
#to know the common 3 , just change in count.update the number
```

### Question 4. What product sold the most? Why do you think it sold the most?

```
product_group = all_months1.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']


products = [product for product, df in product_group]
plt.bar(products , quantity_ordered)
plt.xlabel('product')
plt.ylabel('Quantity Ordered')
plt.xticks(products, rotation = 'vertical',size = 9)
plt.show()
```

```
#Why are some higher than others
#look for some correlation

prices = all_months1.groupby('Product').mean()['Price Each']
#adding second y axis on matplotlib chart

fig,ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(products , quantity_ordered, color ='g')
ax2.plot(products,prices, 'b-')


ax1.set_xlabel('Product name')
ax1.set_ylabel('Quantity Ordered', color = 'g')
ax2.set_ylabel('price ($)', color = 'b')
ax1.set_xticklabels(products, rotation = 'vertical', size = 8)
plt.show()

```
